```python
from sklearn.tree import DecisionTreeRegressor

x_treino, x_teste, y_treino,  y_teste = train_test_split(x, y, test_size = 0.25) 

modelo_arvore = DecisionTreeRegressor(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)

predicoes_matematica_arvore = modelo_arvore.predict(x_teste)

mean_squared_error(y_teste, predicoes_matematica_arvore)
```

```python
from sklearn.model_selection import cross_validate

modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultado = cross_validate(modelo_arvore, x, y, scoring= 'neg_mean_squared_error')
resultado['test_score']*-1
```

```python
from sklearn.model_selection import cross_validate

modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultado = cross_validate(modelo_arvore, x, y, scoring= 'neg_mean_squared_error')
media = (resultado['test_score']*-1).mean()
desvio_padrão = (resultado['test_score']*-1).std()

lim_inferior = media - (2 * desvio_padrão)
lim_superior = media + (2 * desvio_padrão)

print(f"Intervalo de confiança {lim_inferior} - {lim_superior}")
```

```python
def calcula_mse(resultados):
  media = (resultado['test_score']*-1).mean()
  desvio_padrão = (resultado['test_score']*-1).std()

  lim_inferior = media - (2 * desvio_padrão)
  lim_superior = media + (2 * desvio_padrão)

  print(f"Intervalo de confiança {lim_inferior} - {lim_superior}")
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
import numpy as np

SEED = 1232
np.random.seed(SEED)

partes = KFold(n_splits = 10, shuffle=True)
modelo_arvore = DecisionTreeRegressor(max_depth=3)
resultado = cross_validate(modelo_arvore, x, y, cv = partes, scoring= 'neg_mean_squared_error')
calcula_mse(resultado)
```

```python
def regressor_arvore(nivel):
  SEED = 1232
  np.random.seed(SEED)
  partes = KFold(n_splits = 10, shuffle=True)
  modelo_arvore = DecisionTreeRegressor(max_depth=nivel)
  resultado = cross_validate(modelo_arvore, x, y, cv = partes, scoring= 'neg_mean_squared_error', return_train_score=True)
  print(f' Treino = {(resultado["train_score"]*-1).mean()} | Teste = {(resultado["test_score"]*-1).mean()}')
```

```python
for i in range(1,21):
  regressor_arvore(i)
```
